```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
df['Dogum_Tarihi'] = pd.to_datetime(df['Dogum_Tarihi'])

today = pd.to_datetime('today')

df['Yas'] = today.year - df['Dogum_Tarihi'].dt.year

mask = (today.month < df['Dogum_Tarihi'].dt.month) | \
       ((today.month == df['Dogum_Tarihi'].dt.month) & (today.day < df['Dogum_Tarihi'].dt.day))

df['Yas'] -= mask.astype(int)
```

```python
df[['Cinsiyet', 'Il', 'Alerjilerim', 'Kronik Hastaliklarim',
    'Baba Kronik Hastaliklari', 'Anne Kronik Hastaliklari',
    'Kiz Kardes Kronik Hastaliklari', 'Erkek Kardes Kronik Hastaliklari']] = df[
    ['Cinsiyet', 'Il', 'Alerjilerim', 'Kronik Hastaliklarim',
     'Baba Kronik Hastaliklari', 'Anne Kronik Hastaliklari',
     'Kiz Kardes Kronik Hastaliklari', 'Erkek Kardes Kronik Hastaliklari']].fillna('Unknown')
```

```python
blood_groups = ['A RH+', 'A RH-', 'B RH+', 'B RH-', 'AB RH+', 'AB RH-', 'O RH+', 'O RH-']

missing_indices = df['Kan Grubu'].isnull()

df.loc[missing_indices, 'Kan Grubu'] = np.random.choice(blood_groups, size=missing_indices.sum())
```

```python
import pandas as pd
import numpy as np

mean_kilo = df['Kilo'].mean()
std_kilo = df['Kilo'].std()

mean_boy = df['Boy'].mean()
std_boy = df['Boy'].std()

df.loc[df['Kilo'].isnull(), 'Kilo'] = (
    np.random.normal(loc=mean_kilo, scale=std_kilo * 0.1, size=df['Kilo'].isnull().sum()).round(1)
).astype(float)

df.loc[df['Boy'].isnull(), 'Boy'] = (
    np.random.normal(loc=mean_boy, scale=std_boy * 0.1, size=df['Boy'].isnull().sum()).round(1)
).astype(float)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(df['Kilo'], bins=20, color='skyblue', edgecolor='black')
plt.title('Kilo ablonu')
plt.xlabel('Kilo')
plt.ylabel('Kii Says')

plt.subplot(1, 2, 2)
plt.hist(df['Boy'], bins=20, color='salmon', edgecolor='black')
plt.title('Boy ablonu')
plt.xlabel('Boy')
plt.ylabel('Kii Says')

plt.tight_layout()
plt.show()
```
